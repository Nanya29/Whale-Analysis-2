# Custom portfolio 

1. Choose 3-5 custom stocks with at last 1 year's worth of historic prices and create a DataFrame of the closing prices and dates for each stock.

```python
# import yfinance 
import yfinance as yf 

# define function to input dataset
def get_stocks(name, start_date, end_date):
    stock_data =  yf.download(name,start = start_date, end = end_date)
    return stock_data['Close']
```

```python
# Read stock 1 
name = "INTU"
start_date = "2015-11-11"
end_date = "2018-11-11"

INTU = get_stocks(name, start_date, end_date)
INTU
```

```python
# Read stock 2
name = "CRM"
start_date = "2015-11-11"
end_date = "2018-11-11"

CRM = get_stocks(name, start_date, end_date)
CRM
```

```python
# Read stock 3
name = "PGR"
start_date = "2015-11-11"
end_date = "2018-11-11"

PGR = get_stocks(name, start_date, end_date)
PGR
```

```python
CRM.dtypes
```

```python
# Combine all stocks in a single dataframe
custom_stocks = pd.concat([INTU,CRM,PGR],axis = "columns", join = "inner")
custom_stocks
```

```python
# Drop nulls
custom_stocks.dropna(inplace= True)
```

```python
#Count number of nulls
custom_stocks.isnull().sum()
```

```python
#Rename column 
columns=["INTU_Close", "CRM_Close", "PGR_Close"]
custom_stocks.columns = columns
custom_stocks.head()
```

```python
# Calculate daily returns and drop nulls from data frames
custom_stocks_returns = custom_stocks.pct_change()

# Drop nulls from 'custom_stocks_returns'
custom_stocks_returns.dropna( inplace = True)

# Read 'custom_stocks_returns'
custom_stocks_returns.head()

```

```python
# Rename Columns
columns=["INTU_return", "CRM_return", "PGR_return"]
custom_stocks_returns.columns = columns
custom_stocks_returns.head()
```

2. Calculate the weighted returns for the portfolio assuming an equal number of shares for each stocks


```python
# set weights
weights = [1/3, 1/3, 1/3]

# Calculate portfolio return
portfolio_returns = custom_stocks_returns.dot(weights)

# display sample data
portfolio_returns.sample(5)
```

3. Join the calculated portfolio returns dataframe to original returns dataframe

```python
# Use the concat function to join custom portfolio and original portfolio
all_returns = pd.concat([custom_stocks_returns,whale_analysis_data],axis = "columns", join = "inner")

# drop NaNs
all_returns.dropna(inplace = True)

# Display dataframe
all_returns
```

```python
# Calculate the Annualized standard Deviation for 'all_returns'
all_returns.std() * np.sqrt(252)
```

```python
# calculate and plot the rolling standard deviation with a 21 day window
all_returns.rolling(window=21).std().plot(figsize= (10,5))
```

```python
# Calculate and plot correlation
# Calculate the correlation
all_returns_correlation = all_returns.corr()
all_returns_correlation.plot()
```

```python
# Calculate  a 60-day rolling covariance of a single portfolio from the given portfolios  
INTU_rel = all_returns['INTU_return'].rolling(window=60).cov(all_returns['Portfolio_return'])
INTU_rel
```

```python
# calculate a 60 day rollinng variance of the 'Portfolio return'
Portfolio_return_variance = all_returns['Portfolio_return'].rolling(window=60).var()
Portfolio_return_variance
```

```python
# Calculate and plot beta
rolling_beta_2 = INTU_rel / Portfolio_return_variance
rolling_beta_2.plot()
```

```python
# calculate the annualized sharpe ration using the daily returns
sharpe = (all_returns.mean() *252)/ (all_returns.std() * np.sqrt(252) )
sharpe
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe.plot(kind= 'bar')
```
